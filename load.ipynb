```python
import matplotlib.pyplot as plt
```

```python
from modelnet10_rgbd_dataset import ModelNet10RGBDDataset, make_dataloaders

# Single dataset
# ds = ModelNet10RGBDDataset(root="./ModelNet10", split="train")

# Or both splits as DataLoaders
train_loader, test_loader = make_dataloaders("./ModelNet10", batch_size=1, num_workers=0)

batch = next(iter(train_loader))
```

```python
batch['rgb'].shape
```

```python
batch['voxels'].shape
```

```python
rgb = batch["rgb"].float()
depth = batch["depth"].float() / 255.0 
voxels = batch["voxels"][0]

fig, axes = plt.subplots(3, 2, figsize=(5, 10))

for i in range(1):
    # RGB
    axes[0, i].imshow(rgb[i].permute(1, 2, 0).numpy())
    axes[0, i].set_title(f"RGB {i}")
    
    # Depth
    axes[1, i].imshow(depth[i, 0].numpy(), cmap="plasma")
    axes[1, i].set_title(f"Depth {i}")
    
    # Voxel slice (middle)
    axes[2, i].imshow(voxels[0, :, :, 16].numpy(), cmap="gray")
    axes[2, i].set_title("Voxel Z/2")

for ax in axes.flat:
    ax.axis("off")

plt.tight_layout()
plt.show()
```
